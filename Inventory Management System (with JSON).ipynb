```python
import json
import time

sales=""
fd=open("chocolates.json","r")
js=fd.read()
fd.close()

#Converting json string back to dicitionary, dictionary storing the details of the chocolates
chocolates=json.loads(js)


#Chocolate menu
print("--------------------CHOCOLATES--------------------")
for i in chocolates.keys():
    print(i,":",chocolates[i]['Name'],chocolates[i]['Price'],chocolates[i]['Qty'])
print("--------------------------------------------------")

#taking user input
user_name=str(input("Enter your name: "))
user_email=str(input("Enter email: "))
user_phone=str(input("Enter mobile number: "))
user_cid=str(input("Enter chocolate id: "))
user_cqty=int(input("Enter quantity: "))
print("--------------------------------------------------")

#Generating bill
print("                   USER DETAILS                   ")

#checking if required quantity is less than or equal to available quantity
if(user_cqty<=chocolates[user_cid]['Qty']):
    print("-----------------------BILL-----------------------")
    print("Name:                       ",chocolates[user_cid]['Name'])
    print("Price:                      ",chocolates[user_cid]['Price'])
    print("Quantity:                   ",user_cqty)
    print("--------------------------------------------------")
    
    total_amnt=user_cqty*chocolates[user_cid]['Price']
    if(total_amnt>=1000):
        print("Congrats! A 10% discount is applied for all the purchases greater than or equal to Rs.1000")
        total_amnt=total_amnt-(total_amnt/10)
    print("Amount to be paid (Rs):     ",total_amnt)
    

    chocolates[user_cid]['Qty']=chocolates[user_cid]['Qty']-user_cqty
    
    sales=user_name+","+user_email+","+user_phone+","+user_cid+","+chocolates[user_cid]['Name']+","+str(chocolates[user_cid]['Price'])+","+str(user_cqty)+","+str(total_amnt)+","+time.ctime()+"\n"
else:
    print("Sorry, the available quantity is: ",chocolates[user_cid]['Qty'])
    print("To proceed with the whole available quantity, press Y/y: ")
    
    ch=str(input())
    #if user wants to proceed with the whole available quantity
    if(ch=='Y' or ch=='y'):
        print("-----------------------BILL-----------------------")
        print("Name:                       ",chocolates[user_cid]['Name'])
        print("Price:                      ",chocolates[user_cid]['Price'])
        print("Quantity:                   ",chocolates[user_cid]['Qty'])
        print("--------------------------------------------------")
        
        total_amnt=chocolates[user_cid]['Qty']*chocolates[user_cid]['Price']
        if(total_amnt>=1000):
            print("Congrats! A 10% discount is applied for all the purchases greater than or equal to Rs.1000")
            total_amnt=total_amnt-(total_amnt/10)
            
        print("Amount to be paid (Rs):     ",total_amnt)
        if(total_amnt>=1000):
            print("Congrats! A 10% discount is applied for all the purchases greater than or equal to Rs.1000")
            total_amnt=total_amnt-(total_amnt/10)
        
        sales=user_name+","+user_email+","+user_phone+","+user_cid+","+chocolates[user_cid]['Name']+","+str(chocolates[user_cid]['Price'])+","+str(chocolates[user_cid]['Qty'])+","+str(total_amnt)+","+time.ctime()+"\n"

        chocolates[user_cid]['Qty']=0
    else:
        print("Thanks")
#updating/converting dictionary to JSON format
json_string=json.dumps(chocolates)
fd=open("chocolates.json","w")
fd.write(json_string)
fd.close()

#generating sales
fd=open("Sales.txt","a")
fd.write(sales)
fd.close()


print(" ")
print("-------------Thankyou for your order!-------------")
print("Inventory updated...")
print(" ")
```
